## Data Cleaning and Preprocessing

### 4. Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix. 

- Note: When the record/entry has "M"(indicates millions) as suffix you need to multiply that value with 1000000
- When the record/entry has "K"(indicates thousands) as suffix you need to multiply that value with 1000

```python
cols = ['Value', 'Wage', 'Release Clause']
df[cols] = df[cols].replace({'€': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float)
df.head()
```

### 5. Convert the column "Joined" into integer data type with keeping only the year. 

```python
df['Joined'].head()
```

```python
df['Joined'] = pd.to_datetime(df['Joined']).dt.year.astype('Int64')
print('Datatype - ', df['Joined'].dtype)
df[['Name','Joined']].head()
```

### 6. Convert the column "Contract Valid Until" to pandas datetime type.

```python
df['Contract Valid Until'].head()
```

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'], format='mixed')
df[['Name','Contract Valid Until']].head()
```

### 7. The column "Height" is in inches with a quotation mark, Convert to float with decimal points. 

```python
df['Height'].head()
```

```python
df['Height'] = df['Height'].str.replace('\'','-')
```

```python
df['Height'].head()
```

```python
feet = df['Height'].str.split('-', n=1).str[0].astype(float)
inches = df['Height'].str.split('-', n=1).str[1].astype(float)

df['Height'] = (feet+(inches/12)).round(decimals=2)
df[['Name', 'Height']].head()
```

### 8. The column "Weight" has the suffix as lbs, remove the suffix and convert to float.

```python
df['Weight'].head()
```

```python
df['Weight'] = df['Weight'].str.removesuffix('lbs').astype(float)
print('Datatype - ', df['Weight'].dtype)
df[['Name', 'Weight']].head()
```

### 9. Check for the percentage of missing values and impute them with appropriate imputation techniques. 
Finding the Missing/Null Values -

```python
missing_value = (df.isnull().sum() * 100 / len(df)).round(2)
df_missing_value = pd.DataFrame(missing_value)
df_missing_value
```

- From the above we infer that the column "Loaned From" is having missing values which is greater than 90%.
- Therefore, we drop the column "Loaned From" in the dataset.

```python
df.drop('Loaned From', axis=1, inplace=True)
print('Column "Loaned From" is dropped.')
```

**Columns with Missing Numerical Values -**

```python
missing_data = df.isnull().sum()/len(df)
data_index = [i  for i in missing_data[missing_data>0].index]

numerical = [i for i in data_index 
           if i in df.columns and  df[i].dtype!=object]
df_numerical = pd.DataFrame(numerical)
df_numerical
```

**Replacing Missing Numerical Values with the Median Value.**

```python
for j in numerical:
    df[j]=df[j].fillna(df[j].median())
    
print('Missing Numerical Values Replaced with the Median')
```

**Columns with Missing Categorical Values -**

```python
categorical = [i for i in data_index      
           if i in df.columns and  df[i].dtype==object]
df_categorical = pd.DataFrame(categorical)
df_categorical
```

**Replacing Missing Categorical Values with Mode.**

```python
for j in categorical:
    df[j]=df[j].fillna(df[j].mode()[0])
    
print('Missing Categorical Values Replaced with Mode')
```

**Re-Checking for Missing Value.**

```python
missing_data1 = df.isnull().sum()
df_missing_data1 = pd.DataFrame(missing_data1)
df_missing_data1.reset_index()
```

### 21. What is the average wage one can expect to pay for the top 5 players in every position? 

```python
positions = df['Position'].unique()
result = []

for pos in positions:
    top5 = df[df['Position'] == pos].sort_values(by='Overall', ascending=False).head(5)
    avg_wage = top5['Wage'].mean()
    result.append({'Position': pos, 'Average Wage': avg_wage})

result_df = pd.DataFrame(result)

result_df
```
